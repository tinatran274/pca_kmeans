```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Mall_Customers.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
df.drop("CustomerID", axis=1, inplace=True)
```

```python
df.head()
```

```python
for row in df:
    data = df.values.tolist()
print(data)
```

```python
# Standardization
for j in range(len(data[0])):
    sum = 0
    for i in range(len(data)):
        sum += data[i][j]
    average = sum/len(data)
    for i in range(len(data)):
        data[i][j] -= average
print(np.array(data))
```

```python
# Covariance Matrix Computation
data = np.array(data)
covdata = np.cov(data.T)
print(covdata)
```

```python
#Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covdata)
print('Eigen values:\n', eigenvalues)
print('Eigen vector:\n', eigenvectors)
```

```python
#Sort the eigenvalues, Eigenvectors in descending order
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]
print(eigenvalues)
print(eigenvectors)
```

```python
n_components = 2
Uk = []
for i in range(len(eigenvectors)):
    temp = []
    for j in range(n_components):
        temp.append(eigenvectors[i][j])
    Uk.append(temp)
    
Uk = np.array(Uk)
print(Uk.T)
```

```python
new_data = np.dot(Uk.T,data.T)
print(new_data.T)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_coords = new_data.T[:, 0]
y_coords = new_data.T[:, 1]

# Create the plot
plt.scatter(x_coords, y_coords, c='blue', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Data Points')
plt.legend()
plt.grid(True)
plt.show()
```
